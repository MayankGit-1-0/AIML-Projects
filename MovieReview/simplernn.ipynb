```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
## Load he imdb dataset

max_features=10000 ##vocab size
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)

## Print the shape of the data

print(f'training data shape:{X_train.shape},Training labels shape: {y_train.shape}')
print(f'training data shape:{X_train.shape},Training labels shape: {y_test.shape}')
```

```python
## Inspect a sample review and its label

sample_review=X_train[0]
sample_label=y_train[0]

print(f'Sample review (as integers):{sample_review}')
print(f'Sample label:{sample_label}')
```

```python
### mapping of words index bacl to words(for understanding)
word_index=imdb.get_word_index()
word_index
reverse_word_index={value: key for key, value in word_index.items()}
```

```python
decoded_review=' '.join([reverse_word_index.get(i-3,"?") for i in sample_review])
decoded_review
```

```python
from tensorflow.keras.preprocessing import sequence

max_len=500

X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test=sequence.pad_sequences(X_test,maxlen=max_len)
X_train
```

```python
## Train Simple RNN

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense

model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len))
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
## create an instance of EarlyStopping Callback

from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
## Train model with early stopping
model.fit(
    X_train,y_train,epochs=1,batch_size=32,
    validation_split=0.2,
    callbacks=[earlystopping]
)
```

```python
## Save model file

model.save('simple_rnn_imdb.h5')
```
